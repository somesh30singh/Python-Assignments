```python
def find_largest_num(numb : list[float]) -> float:
    """This function will return largest number
    args:
        int: (numb :list(float)) input must be int or float
    return:
          it will return largest number
    raise:
         TypeError: input must be float or int
         ValueError: Input should not be empty
    """

    if not isinstance(numb,list):
        raise TypeError("input must be float or int")
    if not numb:
        raise ValueError("input should not be empty")
    for item in numb:
        if not isinstance(item,(float,int)):
            raise TypeError("input must be numeric")
    largest = numb[0]
    for num in numb:
        if num > largest:
            largest = num
    return largest

if __name__=="__main__":
    try:
        user_input = input("Enter Your Number , sererate by space")
        num = [float (item)for item in user_input.split()]
        result = find_largest_num(num)
        print(f"largest number{result}")
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as error:
        print(error)
```
