```python
# REVERSE STRING
string = ("123456")
rev = ""
for item in string:
    rev = item + rev
print(rev)
```

```python
def reverse_string(string1: str) -> str:
    """This function will return reverse string
    args:
        we will use for loop and exception handling to run our code
    return:
          in return it will provide reverse string
    """

    if not isinstance(string1,str):
        raise TypeError("Input must be string")
    rev = ""
    for item in string1:
        rev = item + rev
    return rev

if __name__=="__main__":
    try:
        user_input = input("ENTER YOUR STRING")
        result = reverse_string(user_input)
        print(f" Reverse string {result}")
    except TypeError as error:
        print(error)
    except Exception as error:
        print("Something went wrong")
```
